```python
data = pd.read_csv(r"C:\Users\Chris\OneDrive\Documents\Python Projects\BTC Price Prediction\BTC-USD.csv")

print(data)
```

```python
date_range = pd.date_range(start="2022-1-1", end="2023-08-15")
price = data["Close"].values

ts = pd.DataFrame(price, index=date_range)

plt.plot(ts)
plt.title("Price of Bitcoin from Jan 01, 2022 to August 15, 2023")
plt.legend(['Daily Price of Bitcoin'])
plt.xlabel("Date")
plt.ylabel("Price in USD")
plt.xticks(rotation=20)  # Adjust the rotation angle as needed
plt.show()
```
